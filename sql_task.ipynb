__Query3:insert__

insert values into table

```python
import csv
con=sql.connect(host="localhost",user="root",password="root",port=3306,database="employee")
cursor=con.cursor()
with open(r'C:\Users\ineuron\19.2.MySQL\glass.data',"r") as data :
    next(data)
    data_csv = csv.reader(data, delimiter= "\n")
    print(data_csv)
    for i in enumerate(data_csv):
        print(i)
        for j in i[1] :
            cursor.execute('insert into employee.GlassData values ({data})'.format(data=(j)))
    print("all the data inserted ")
con.commit()
```

__Query 12: update: update table name  set column name= coumn name+new value__

__Query 13:delete:delete from table name where condtion__

delete one row from table where condition match


```python
con=sql.connect(host="localhost",user="root",password="root",port=3306,database="employee")
cursor=con.cursor()
cursor.execute("delete from GlassData where Index_Number=5")
cursor.execute("select * from GlassData")
cursor.scroll(0,mode="absolute")#----->when row 5 deleted cursor goes on 6th row so for fetching all row we havw to go 1st position
rows=cursor.fetchall()
print(rows)
con.commit()
con.close()

```
